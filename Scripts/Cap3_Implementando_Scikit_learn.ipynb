```python
import pandas as pd

# Carregando o dataset diretamente da internet
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt"

# Definindo os nomes das colunas conforme a descrição do dataset
colunas = [
    "area",
    "perimetro",
    "compacidade",
    "comprimento_nucleo",
    "largura_nucleo",
    "coef_assimetria",
    "comprimento_sulco",
    "classe"  # 1 = Kama, 2 = Rosa, 3 = Canadian
]

# Lendo o dataset
df = pd.read_csv(url, sep='\s+', header=None, names=colunas)

# Exibindo as primeiras 5 linhas
df.head()
```

```python
from scipy import stats
import numpy as np

# Calculando o z-score
z_scores = np.abs(stats.zscore(df.iloc[:, :-1]))  # ignora a coluna 'classe'
df_sem_outliers = df[(z_scores < 3).all(axis=1)]
```

```python
from sklearn.preprocessing import StandardScaler

# Criando o scaler
scaler = StandardScaler()

# Separando X e y
X = df.drop("classe", axis=1)
y = df["classe"]

# Aplicando padronização
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Separando em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# Inicializando os modelos
model_knn = KNeighborsClassifier()
model_svm = SVC()
model_rf = RandomForestClassifier(random_state=42)

# Treinando os modelos
model_knn.fit(X_train, y_train)
model_svm.fit(X_train, y_train)
model_rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Função para avaliar modelos
def avaliar_modelo(nome, modelo):
    y_pred = modelo.predict(X_test)
    print(f"=== {nome} ===")
    print("Acurácia:", accuracy_score(y_test, y_pred))
    print("Relatório de Classificação:\n", classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Matriz de Confusão - {nome}")
    plt.xlabel("Previsto")
    plt.ylabel("Real")
    plt.show()

avaliar_modelo("KNN", model_knn)
avaliar_modelo("SVM", model_svm)
avaliar_modelo("Random Forest", model_rf)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 4],
    'min_samples_leaf': [1, 2]
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='accuracy')
grid_rf.fit(X_train, y_train)

best_rf = grid_rf.best_estimator_
print("Melhores parâmetros:", grid_rf.best_params_)
```

```python
param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

grid_svm = GridSearchCV(SVC(), param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_train, y_train)

best_svm = grid_svm.best_estimator_
print("Melhores parâmetros:", grid_svm.best_params_)
```

```python
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]  # p=1 (Manhattan), p=2 (Euclidiana)
}

grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5, scoring='accuracy')
grid_knn.fit(X_train, y_train)

best_knn = grid_knn.best_estimator_
print("Melhores parâmetros:", grid_knn.best_params_)
```

```python
avaliar_modelo("KNN", best_knn)
avaliar_modelo("SVM", best_svm)
avaliar_modelo("Random Forest", best_rf)
```
